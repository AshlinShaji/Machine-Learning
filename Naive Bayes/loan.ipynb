```python
df['Married']=df['Married'].fillna(df['Married'].mode()[0])
```

```python
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
```

```python
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
df['Term']=df['Term'].fillna(df['Term'].mode()[0])
```

```python
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

```python
df=df.drop(['Gender'],axis=1)
```

```python
df.tail()
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Married']=lab.fit_transform(df['Married'])
df['Dependents']=lab.fit_transform(df['Dependents'])
df['Education']=lab.fit_transform(df['Education'])
df['Self_Employed']=lab.fit_transform(df['Self_Employed'])
df['Area']=lab.fit_transform(df['Area'])
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.30,random_state=1)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit_transform(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
lab=['Y','N']
cmd=ConfusionMatrixDisplay(cm,display_labels=lab)
cmd.plot()
```
