```python
df_1.dtypes
```

```python
#string replace
df_1['Mileage']=df_1['Mileage'].str.replace('km/kg','')
df_1['Mileage']=df_1['Mileage'].str.replace('kmpl','')
df_1['Engine']=df_1['Engine'].str.replace('CC','')
df_1['Power']=df_1['Power'].str.replace('bhp','')
df_1['Mileage']=df_1['Mileage'].str.replace('null','0')
df_1['Engine']=df_1['Engine'].str.replace('null','0')
df_1['Power']=df_1['Power'].str.replace('null','0')
```

```python
df_1['Power'].unique()
```

```python
df_1
```

```python
df_1.dtypes
```

```python
#dtata type change
df_1['Engine']=df_1['Engine'].astype(float)
df_1['Mileage']=df_1['Mileage'].astype(float)
df_1['Power']=df_1['Power'].astype(float)
```

```python
df_1.dtypes
```

```python
#Mileage,Engine,Power.  ====>null. ===>0
#numerical value. 0
#missing value


df_1.loc[df_1.Engine==0,'Engine']=np.nan
df_1.loc[df_1.Power==0,'Power']=np.nan
df_1.loc[df_1.Mileage==0,'Mileage']=np.nan
```

```python
df_1.isna().sum()
```

```python
df_1['Mileage']=df_1['Mileage'].fillna(df_1['Mileage'].mean())
df_1['Engine']=df_1['Engine'].fillna(df_1['Engine'].mean())
df_1['Power']=df_1['Power'].fillna(df_1['Power'].mean())
df_1['Seats']=df_1['Seats'].fillna(df_1['Seats'].mean())
```
